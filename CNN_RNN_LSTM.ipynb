# **CNN**

-CNN은 주로 이미지나 공간적인 데이터를 처리하는 데 특화된 모델이다.

-이미지 전체를 한꺼번에 보는 것이 아니라, 필터(Filter)을 사용하여 이미지의 부분적인 특징(세로선, 가로선, 질감 등)을 추출한다.

-컨볼루션 신경망(CNN)은 동물의 조직에서 영감을 얻어서 만들어진 신경망이다.

-CNN에서 뉴런의 수용 공간은 다른 뉴론들과 약간 겹치게 된다.

-데이터의 공간적인 구조를 유지하며 학습하기 때문에, 이미지 인식에서 압도적인 성능을 보인다.

-컨볼루션 신경망은 영상 및 비디오 인식, 추천시스템 및 자연 언어 처리 분야에서 폭넓게 응용되고 있다.

-CNN 신경망에서는 하위 레이어의 노드들과 상위 레이어의 노드들이 부분적으로만 연결되어 있다. 따라서 복잡도가 낮아지고 과대 적합에 빠지지 않는다.

-컨볼루션은 주변 화소값들에 가중치를 곱해서 더한 후에 이것을 새로운 화소값으로 하는 연산이다.

-폴링이란 서브 샘플링이라고도 하는 것으로 입력 데이터의 크기를 줄이는 것이다.


# **RNN**

-순서가 있는 데이터(Sequence Data)나 시계열 데이터를 처리하기 위해 고안된 모델이다.

-'기억'을 가지고 있다. 이전 단계의 출력값이 다시 입력값으로 들어가는 순환 구조를 가진다.

-현재의 입력과 이전 상태의 기억을 함께 계산하여 현재의 상태를 만든다.

-입력 문장의 길이에 상관없이 모델을 적용할 수 있는 유연함이 있다.

-RNN(Recurrent Neural Network)은 시계열 또는 자연어와 같은 시스 데이터를 모델링하는 데 강력한 신경망이다.

-RNN은 시계열 데이터나 시퀀스 데이터를 잘 처리할 수 있다.

-RNN은 시간을 통해 오차를 역전파하여 학습한다.(BPTT)

-RNN은 음악 생성이나 기계 번역 등에 사용할 수 있고 주가 예측이나 텍스트 처리, 기후 예측과 같은 분야에도 사용할 수 있다.

-단) 입력 데이터가 길어질수록 앞부분의 정보를 잊어버리는 장기 의존성 문제가 발생한다.

-단) 역전파 과정에서 가중치를 계속 곱하다 보니 기울기가 너무 작아지거나 너무 커지는 문제가 발생한다.



# **LSTM**

-LSTM(Long short-term memory)는 기존 RNN을 훈련한 때 발생할 수 있는 그래디언트 소실 문제를 해결하기 위해 개발되었다.

-정보를 잊을지, 저장할지, 출력할지를 결정하는 게이트 시스템을 도입했다.

-3가지 게이트: Forget Gate(과거의 정보 중 무엇을 버릴지 결정), Input Gate(현재의 정보 중 무엇을 저장할지 결정), Output Gate(어떤 부분을 다음 단계로 내보낼지 결정)

-Cell State를 통한 덧셈 연산 위주의 업데이트 덕분에 RNN에 비해 기울기 소실 문제가 현저히 적다.

-긴 문장이나 긴 시계열 데이터에서도 중요한 정보를 오랫동안 유리할 수 있어 RNN보다 훨씬 널리 쓰인다.

-긴 문장 번역, 챗봇, 작곡 등에 활용할 수 있다.

-RNN보다 구조가 복잡하여 연산량이 많고 학습 시간이 더 오래 걸린다는 단점이 있다.
