```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
train_images = train_images.reshape((60000, 28, 28, 1)) # мқҙлҜём§Җ лҶ’мқҙ, мқҙлҜём§Җ л„Ҳл№„, мұ„л„җ к°ңмҲҳ
test_images = test_images.reshape((10000, 28, 28, 1))
```

```python
# н”Ҫм…Җ к°’мқ„ 0~1 мӮ¬мқҙлЎң м •к·ңнҷ”
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
# м»ЁлІҢлЈЁм…ҳ мӢ кІҪл§қ мғқм„ұ
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1))) # мҲ«мһҗ мқҙлҜём§Җ лӘЁм–‘мқё (28,28,1) нҠңн”Ңмқ„ input_shapeмқҳ л§Өк°ңліҖмҲҳлЎң м „лӢ¬
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# л§Ҳм§Җл§үм—җ dense л Ҳмқҙм–ҙ м¶”к°Җ
model.add(layers.Flatten()) # dense л Ҳмқҙм–ҙлҠ” 1м°Ёмӣҗ н…җм„ңлҘј мһ…л Ҙ л°ӣмқҢ -> нҸүнғ„нҷ”
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax')) # мҲ«мһҗ мқҙлҜём§ҖлҠ” 10к°ң мң нҳ•
```

```python
# лӘЁлҚё м¶ңл Ҙ
model.summary() # Conv2D л Ҳмқҙм–ҙм—җм„ңлҠ” мқҙлҜём§Җмқҳ нҒ¬кё°к°Җ 2л§ҢнҒј(нҢЁл”©) мӨ„м–ҙл“Өкі , MaxPooling2 л Ҳмқҙм–ҙм—җм„ңлҠ” м Ҳл°ҳмңјлЎң мӨ„м–ҙл“Ұ
```

```python
# м»ҙнҢҢмқјкіј нӣҲл Ё
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # sparse_categorical_crossentropy -> нӣҲл Ё мғҳн”Ңмқҳ м¶ңл Ҙмқ„ мӣҗ-н•« мқёмҪ”л”©мңјлЎң л§Ңл“Өм§Җ м•Ҡм•„лҸ„ лҗЁ
model.fit(train_images, train_labels, epochs=5)
```

```python
# м „мІҙ кІ°кіј мҳҲмёЎ
y_pred = model.predict(test_images)
y_pred_classes = np.argmax(y_pred, axis=1)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(test_labels, y_pred_classes))

print("\nClassification Report:")
print(classification_report(test_labels, y_pred_classes))
```

```python
# к°„лӢЁн•ҳкІҢ кө¬н•ҳкі  мӢ¶лӢӨл©ҙ(мҶҗмӢӨкіј м •нҷ•лҸ„л§Ң) evaluate()
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("н…ҢмҠӨнҠё м •нҷ•лҸ„:", test_acc)
```
